```python
import pandas as pd
from datetime import date
```

```python
df = pd.read_csv("Final_Datasets/final_dataset.csv")
```

```python
df
```

```python
df.columns
```

```python
X = df[['Date', 'Month', 'Year', 'Holidays_Count', 'Days']]
Y = df['NO2']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.30)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score,mean_absolute_error

models = {

    "Random Forest" : RandomForestRegressor(),

}

for i in range(len(list(models))):
    model = list(models.values())[i]
    model.fit(X_train,Y_train)
    Y_pred = model.predict(X_test)

    acc = r2_score(Y_test,Y_pred)
    mae = mean_absolute_error(Y_test,Y_pred)
    print(f"R2 value for {list(models.keys())[i]} : ", acc*100)
    print(f"MAE value for {list(models.keys())[i]} : ", mae)
    print("-"*50) # Just to make a boundary
```
